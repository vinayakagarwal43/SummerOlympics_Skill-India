# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("summer.csv")
df.head()
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
d2=[]

for sport in df['Sport'].unique():
    data = df[df['Sport']==sport]
    for medal in data['Medal'].unique():
        if medal=='Gold':
            d2.append([sport, len(data[data['Medal']==medal])])

pd.DataFrame(d2,columns = ['sport','golds']).sort_values(by='golds', ascending=False).plot(x='sport', y='golds', kind='bar', figsize = (20,5))

d2[:5][:]
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
d2=[]

for sport in df['Sport'].unique():
    d2.append([sport, len(df[df['Sport']==sport])])
    
pd.DataFrame(d2,columns = ['sport','medals']).sort_values(by='medals', ascending=False).plot(x='sport', y='medals', kind='bar', figsize = (20,5))

d2[:5][:]
```

### 4. Which player has won most number of medals? (Top 5)

```python
d2=[]

for player in df['Athlete'].unique():
    d2.append([player, len(df[df['Athlete']==player])])
    
d2 = sorted(d2,key=lambda l:l[1], reverse=True)
d3 = d2[:30][:]
pd.DataFrame(d3,columns = ['player','medals']).plot(x='player', y='medals', kind='bar', figsize = (20,5))

d3[:5][:]
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
d2=[]

for player in df['Athlete'].unique():
    data = df[df['Athlete']==player]
    for medal in data['Medal'].unique():
        if medal=='Gold':
            d2.append([player, len(data[data['Medal']==medal])])
            
d2 = sorted(d2,key=lambda l:l[1], reverse=True)
d3 = d2[:30][:]
pd.DataFrame(d3,columns = ['player','medals']).plot(x='player', y='medals', kind='bar', figsize = (20,5))

d3[:5][:]
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
d5=[]

for event in df['Event'].unique():
    d5.append([event, len(df[df['Event']==event])])
    
d5 = sorted(d5,key=lambda l:l[1], reverse=True)
d6 = d5[:30][:]
pd.DataFrame(d6,columns = ['event','players']).plot(x='event', y='players', kind='bar', figsize = (20,5))

d6[:5][:]
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
dat1=[]
dat2=[]
d10=[]

for sport in df['Sport'].unique():
    dat1 = df[df['Sport']==sport]
    for medal in dat1['Medal'].unique():
        if medal=='Gold':
            dat2=dat1[dat1['Medal']==medal]
            for sex in dat2['Gender'].unique():
                if sex=='Women':
                    d10.append([sport, len(dat2[dat2['Gender']==sex])])
                
pd.DataFrame(d10,columns = ['sport','FemaleGold']).sort_values(by='FemaleGold', ascending=False).plot(x='sport', y='FemaleGold', kind='bar', figsize = (20,5))

d10[:5][:]
```
